## Import Required Libraries

```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 10)
```

## Load the Datasets

```python
songs = pd.read_csv("../data/raw/data.csv")

artists = pd.read_csv("../data/raw/data_by_artist.csv")

genres = pd.read_csv("../data/raw/data_by_genres.csv")

years = pd.read_csv("../data/raw/data_by_year.csv")

songs_with_genres = pd.read_csv("../data/raw/data_w_genres.csv")
```

```python
datasets = {
    "Songs": songs,
    "Artists": artists,
    "Genres": genres,
    "Years": years,
    "Songs with Genres": songs_with_genres
}

for name, df in datasets.items():
    print(f"{name}: {df.shape}")
```

### Dataset Dimensions

The Spotify project contains five datasets. This step helps us understand how much data is available in each dataset before analysis.

The output shows:

- Number of rows (observations)
- Number of columns (features)
